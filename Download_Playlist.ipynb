```python
# This code is incredible analogous to the before one
# Creator: Amaro, A.
# Description: This code allows you to download a hole YT playlist with/without subtitles
# Date: 17/04/2024

full_description = '''
Hey you! This code is like a 2.0 version of the first code, this 
allows you to download a complete playlist. So, what do you need? 

You only must to get the url of the playlist and create a folder
in your PC (I can do a code to create also the folder 
but I think it's a little easier for you to find the folder if 
you created it!) 

And it's done! This code will get every single video on the 
playlist, download the audio and, if you want it, get the 
transcription in a txt.
''' 

# Here's all the librarys used
from pathlib import Path
from pytube import Playlist
import os 

# You only copy the url and the path and paste it 

#url = "https://youtube.com/playlist?list=PLJicmE8fK0Eg8cBExYSr9r3tzg9ZQo3gD&si=TrGuYbAXl5f7WRxf"
#download_path = Path(r"C:\Users\blume\Downloads\Podcast_Transcrip")

url = str(input("Give me the YT url, pls: "))
download_path = Path(input("\n\nGive the folder path: "))
subtitles = str(input("\n\nDo you want transcriptions: [y/n]?: "))
print("\n\n")

# ----------------- Funciones ------------------
# Esta le da formato al texto, remueve los minutos 
# y los numeros, retorna SOLO el texto
def transcription_format(text):
    new_text = text.split("\n")
    final_text = ''
    
    for i in range(len(new_text)):
        if i%2 == 0 and new_text[i].isdigit() == False:
            final_text += new_text[i] + "\n"  
            
    return final_text
    
# Esta funcion intenta escribrir el archivo txt con el 
# texto. 
def write_file(file_path, final_text):
    try:
        with open(Path(file_path), 'w', encoding = 'utf-8') as archivo:
            archivo.write(str(final_text))
            print("The file was succesfully created! at: ", file_path, "\n\n")
    
    except Exception as e:
        print(f"An error occurred: {e}")
    
# ---------------------------------------------
    
try:
    p = Playlist(url) # Get the playlist
    print("You're about to download: ", p.title) # Shows the playlist
    i = 0
    for video in p.videos:
        try:
            filename = str(video.title) + ".mp4" 

            i += 1 
            total = len(p.videos)

            print("-"*50)
            print("Dowloading [", i, "/", total, "]: "  , filename, "...")
            audio= video.streams.filter(only_audio = True ) # Select only de audio streams
            audio.first().download(download_path) # Get the first stream (best quality) and download it
            print("Done!\n")
            
            if (subtitles == "y"):
                try:
                    transcription = video.captions['en'] # Get the subtitles in English - en
                    text = transcription.generate_srt_captions() # Generate good subtitles

                    #----------- This is just format
                    final_text = transcription_format(text = text)
                    # ----------

                    # Remove windows characters reserved
                    reserved = ["<", ">", ":", ' " ',  "/", "\\", "|", "?", "*"]
                    name = video.title + ".txt"
                    for word_r in reserved:
                        if word_r in name:
                            name = name.replace(word_r, "")
                    
                    #name = str(i) + ".txt"
                    file_path = os.path.join(download_path, name)

                    #-------- Write the txt
                    write_file(file_path = file_path, final_text = final_text)
                    # ---------
            
                except Exception as e:
                    print(f"\n\nSomething went wrong. It's impossible get the correct transcription\n\n {e} \n\n")

        except:
            print("Something also went wrong, I was unable to download it :(")

    print("-"*50)
    print("---> Download completed!") 
                       
        
except: 
    print("Something went wrong. Is correct your URL?")





```
