```python
# So, in YT you can only get the url, so the next step is that you search a YT video and get the url
url = "https://youtu.be/OiKQjezOKXc"

# And the path where you want to save the audio and transcription
from pathlib import Path
download_path = Path(r"C:\Users\blume\Downloads\audios_yt")
```

```python
# And now the fun begins, import the libraries
from pytube import YouTube

try:
    yt = YouTube(url)
    filename = str(yt.title) + ".mp4"
    print("You're about to download: ", yt.title)
    
except: 
    print("Something went wrong. Is correct your URL?")



```

```python
# But we only want the audio, so we're using a stream filter
print("Downloading...")
try:
    audio= yt.streams.filter(only_audio = True ) # Select only de audio streams
    audio.first().download(download_path) # Get the first stream (best quality) and download it
    print("Download completed!")
    
except: 
    print("Something also went wrong, I was unable to download it :(")
```

```python
# And to obtain the subtitle/caption tracks
next = '''
This next step is really trick, cause YT changes its audio format very often, so in order to
get the transcrption in a good format it was necesary to change the library using this steps:

https://github.com/pytube/pytube/issues/1477

After changing the library, everything should be working!
'''

try:
    print("\n\nNow we're trying to get your subtitles... \n\n")
    transcription = yt.captions['en'] # Get the subtitles in English - en
    text = transcription.generate_srt_captions()
    print("Your subtitles corresponds to: \n\n", text)
except:
    print("\n\nSomething went wrong. It's impossible get the correct transcription\n\n")

    
```

```python
# Maybe you want to get a nice format for reading, so:
new_text = text.split("\n")
final_text = ''
for i in range(len(new_text)):
    if i%2 == 0 and new_text[i].isdigit() == False:
        final_text += new_text[i] + "\n"  

print("And your subtitles with a nice format corresponds to: \n\n", final_text)
```

```python
# And the final step is save the subtitles in a .txt
import os
name = yt.title + ".txt"
file_path = (os.path.join(download_path, name)).replace("?", "")
try:
    with open(Path(file_path), 'w') as archivo:
        archivo.write(str(final_text))
    print("The file was succesfully created! at: ", file_path)
except Exception as e:
    print(f"An error occurred: {e}")
```
